```python
class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super(MainWindow, self).__init__()
    
        # qss_file = QtCore.QFile("styles.qss")
        # if qss_file.open(QtCore.QFile.OpenMode.ReadOnly | QtCore.QFile.OpenMode.Text):
        #     stream = QtCore.QTextStream(qss_file)
        #     self.setStyleSheet(stream.readAll())
        #     qss_file.close()
        # self.setStyleSheet(styles)\
        self.mergefiles=[]
        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)
        self.home_page()
        self.upload_mergepdf()
        self.removebtns()
        self.function_btn()
        self.compress_link=''
        self.extract_link=''
        self.extractedtext=''
        self.encrpyt_link=''
        self.dencrpyt_link=''
        
        
    def home_page(self):
        self.ui.statsbtn_15.clicked.connect(lambda: self.ui.stackedWidget.setCurrentWidget(self.ui.page_6))
        self.ui.setupbtn_14.clicked.connect(lambda: self.ui.stackedWidget.setCurrentWidget(self.ui.page))
        self.ui.statsbtn_10.clicked.connect(lambda: self.ui.stackedWidget.setCurrentWidget(self.ui.page_11))
        self.ui.statsbtn_12.clicked.connect(lambda: self.ui.stackedWidget.setCurrentWidget(self.ui.page_7))
        self.ui.setupbtn_11.clicked.connect(lambda: self.ui.stackedWidget.setCurrentWidget(self.ui.page_5))
        self.ui.homesub.clicked.connect(lambda: self.ui.stackedWidget.setCurrentWidget(self.ui.page_3))
        
    def merge_table(self):
        self.ui.tableWidget.setColumnCount(1)
        self.ui.tableWidget.setRowCount(len(list(self.mergefiles)))
        self.ui.tableWidget.setHorizontalHeaderLabels(["File"])
        self.ui.tableWidget.horizontalHeader().setSectionResizeMode(QtWidgets.QHeaderView.Stretch)
        self.ui.tableWidget.verticalHeader().setSectionResizeMode( QtWidgets.QHeaderView.Stretch)
        count=0
        
        
        
        
        for i,l in enumerate(self.mergefiles):
            self.ui.tableWidget.setItem(count,0,QtWidgets.QTableWidgetItem(f'pdf{i+1}'))
            count+=1
        
            
            
            
        
    def upload_mergepdf_btn(self):
        
        dialog=QtWidgets.QFileDialog()
        dialog.exec()
        self.mergefiles.append(dialog.selectedFiles())
        self.merge_page()
        
    
        
    def removebtns(self):
        self.ui.pushButton_11.clicked.connect(lambda:self.remove_merge())  
        self.ui.pushButton_25.clicked.connect(lambda:self.remove_compress())    
        self.ui.pushButton_24.clicked.connect(lambda:self.remove_extract()) 
        self.ui.pushButton_12.clicked.connect(lambda:self.remove_extract())
        self.ui.pushButton_23.clicked.connect(lambda:self.remove_decrypt())
                                              
        
    
    def remove_merge(self):
        self.mergefiles.clear()
        self.merge_page()
        
    def remove_decrypt(self):
        self.dencrpyt_link=''
        self.decypt_page()
        
        
    def upload_mergepdf(self):
        self.ui.pushButton_2.clicked.connect(lambda:self.upload_mergepdf_btn())
        
    def function_btn(self):
        
        
        self.ui.pushButton_3.clicked.connect(lambda:self.real_merge_fuction())
        self.ui.statsbtn_16.clicked.connect(lambda:self.lossless_comp())
        self.ui.statsbtn_17.clicked.connect(lambda:self.remove_dup())
        self.ui.statsbtn_18.clicked.connect(lambda:self.reduce_quality())
        self.ui.statsbtn_19.clicked.connect(lambda:self.remove_img())
        self.ui.pushButton_15.clicked.connect(lambda:self.real_extract())
        self.ui.pushButton_16.clicked.connect(lambda:self.real_encrypt())
        self.ui.pushButton_10.clicked.connect(lambda:self.real_decrypt())
        
    def real_decrypt(self):
        if self.ui.lineEdit_6.text()=='':
            pass
        else:
            if self.ui.lineEdit_2.text()=='':
                pass
            else:
                decrypt_pdf(self.dencrpyt_link,self.ui.lineEdit_6.text(),self.ui.lineEdit_2.text())
                self.ui.lineEdit_6.clear()
                self.remove_decrypt()
        
    
    def lossless_comp(self):
        if self.ui.lineEdit_9.text()=='':
            pass
        else:
            lossless_compression(self.compress_link,self.ui.lineEdit_9.text())
            self.ui.lineEdit_9.clear()
            self.remove_compress()
            
    def remove_dup(self):
        if self.ui.lineEdit_9.text()=='':
            pass
        else:
            compress_remove_duplicates(self.compress_link,self.ui.lineEdit_9.text())
            self.ui.lineEdit_9.clear()
            self.remove_compress()
    
    def reduce_quality(self):
        
        if self.ui.lineEdit_9.text()=='':
            pass
        
        quality=50
        if self.ui.lineEdit_3 != '':
            quality=int(self.ui.lineEdit_3.text())
        else:
            compress_reduceimage_quality(self.compress_link,self.ui.lineEdit_9.text(),quality)
            self.ui.lineEdit_9.clear()
            self.remove_compress()
            
    def remove_img(self):
        if self.ui.lineEdit_9.text()=='':
            pass
        else:
            compress_remove_images(self.compress_link,self.ui.lineEdit_9.text())
            self.ui.lineEdit_9.clear()
            self.remove_compress()
        
    
        
    def real_merge_fuction(self):
        print('cllicked')
        if self.ui.lineEdit_8.text()=='':
            print('NOOOOOOOOO')
            pass
        else:
            merge(self.mergefiles,self.ui.lineEdit_8.text())
            self.ui.lineEdit_8.clear()
            self.remove_merge()       
        
    def merge_page(self):
        
        
       
        
        self.merge_table()
        
    def remove_compress(self):
        self.compress_link=''
        self.compress_page()
    
    
        
    
        
    def upload_compress(self):
        dialog=QtWidgets.QFileDialog()
        dialog.exec()
        self.compress_link=dialog.selectedFiles()
        
           
    def upload_compress_btn(self):
        self.ui.pushButton_13.clicked.connect(lambda:self.upload_compress())
        
    def compress_table(self):
        self.ui.tableWidget_5.setColumnCount(1)
        self.ui.tableWidget_5.setRowCount(len(list(self.mergefiles)))
        self.ui.tableWidget_5.setHorizontalHeaderLabels(["File"])
        self.ui.tableWidget_5.horizontalHeader().setSectionResizeMode(QtWidgets.QHeaderView.Stretch)
        self.ui.tableWidget_5.verticalHeader().setSectionResizeMode( QtWidgets.QHeaderView.Stretch)
        if self.compress_link=='':
            pass
        else:
            self.ui.tableWidget.setItem(0,0,QtWidgets.QTableWidgetItem(f'pdf file'))
            
               
    def compress_page(self):
        self.compress_table()
        
        
    def extract_table(self):
        self.ui.tableWidget_4.setColumnCount(1)
        self.ui.tableWidget_4.setRowCount(len(list(self.mergefiles)))
        self.ui.tableWidget_4.setHorizontalHeaderLabels(["File"])
        self.ui.tableWidget_4.horizontalHeader().setSectionResizeMode(QtWidgets.QHeaderView.Stretch)
        self.ui.tableWidget_4.verticalHeader().setSectionResizeMode( QtWidgets.QHeaderView.Stretch)
        if self.extract_link=='':
            pass
        else:
            self.ui.tableWidget.setItem(0,0,QtWidgets.QTableWidgetItem(f'pdf file'))
            
            
    def upload_extract_btn(self):
        self.ui.pushButton_14.clicked.connect(lambda:self.upload_extract())
            
    def upload_extract(self):
        dialog=QtWidgets.QFileDialog()
        dialog.exec()
        self.extract_link=dialog.selectedFiles()
        self.extract_page()
    
    def remove_extract(self):
        self.extract_link=''
        self.extractedtext=''
        self.extract_page()
        
    def real_extract(self):
        
        if self.ui.lineEdit_5.text()=='':
            pass
        else:
            self.extractedtext=extract_text(self.extract_link,self.ui.lineEdit_5.text())
            
            self.ui.lineEdit_5.clear()
            self.extract_page()
        
        
        
            
        
        
        
        
    
    def extract_page(self):
        self.ui.textEdit.setText(self.extractedtext)
        self.extract_table()
        
        
    def encrypt_table(self):
        self.ui.tableWidget_2.setColumnCount(1)
        self.ui.tableWidget_2.setRowCount(len(list(self.mergefiles)))
        self.ui.tableWidget_2.setHorizontalHeaderLabels(["File"])
        self.ui.tableWidget_2.horizontalHeader().setSectionResizeMode(QtWidgets.QHeaderView.Stretch)
        self.ui.tableWidget_2.verticalHeader().setSectionResizeMode( QtWidgets.QHeaderView.Stretch)
        if self.encrpyt_link=='':
            pass
        else:
            self.ui.tableWidget.setItem(0,0,QtWidgets.QTableWidgetItem(f'pdf file'))
            
    
    def upload_encrypt_btn(self):
        self.ui.pushButton_5.clicked.connect(lambda:self.upload_encrypt())
        
     
    def upload_encrypt(self):
        dialog=QtWidgets.QFileDialog()
        dialog.exec()
        self.encrpyt_link=dialog.selectedFiles()
        self.encypt_page()   
        
    def remove_encrpt(self):
        self.encrpyt_link=''
        self.encypt_page()
    
        
    def real_encrypt(self):
        if self.ui.lineEdit_7.text()=='':
            pass
        else:
            if self.ui.lineEdit.text()=='':
                pass
            else:
                encrypt_pdf(self.encrpyt_link,self.ui.lineEdit_7.text(),self.ui.lineEdit.text())
                self.ui.lineEdit_7.clear()
                self.remove_encrpt()  
        
    def encypt_page(self):
        self.encrypt_table()
        
    
    def decrypt_table(self):
        self.ui.tableWidget_3.setColumnCount(1)
        self.ui.tableWidget_3.setRowCount(len(list(self.mergefiles)))
        self.ui.tableWidget_3.setHorizontalHeaderLabels(["File"])
        self.ui.tableWidget_3.horizontalHeader().setSectionResizeMode(QtWidgets.QHeaderView.Stretch)
        self.ui.tableWidget_3.verticalHeader().setSectionResizeMode( QtWidgets.QHeaderView.Stretch)
        if self.dencrpyt_link=='':
            pass
        else:
            self.ui.tableWidget.setItem(0,0,QtWidgets.QTableWidgetItem(f'pdf file')) 
            
            
        
        
    def decypt_page(self):
        self.decrypt_table()
        
        
        
        
        
        
    
            
          
            
            
        
```
